```python
from PIL import Image

# Display image in colab
image_colored = Image.open("/content/drive/MyDrive/AI ML/Data/eye.png")
display(image_colored)
```

```python

# For colored image

print("Image format: ", image_colored.format)
print ("Image size: ", image_colored.size)
print ("Image mode: ", image_colored.mode)
```

```python
import numpy as np

# For Colored Image

# Convert the pillow image to a numpy array
image_array_colored = np.array(image_colored)

# Display the shape of the numpy array (height, width and chanels)
print("Shape of the image: ", image_array_colored.shape)
```

```python
# Accessing individual channels

import matplotlib.pyplot as plt


red_channel = image_array_colored[:, :, 0]
print(red_channel)
display(red_channel)
plt.imshow(red_channel)
```

```python
# For Green Channel

green_channel = image_array_colored[:, :, 1]
print(red_channel)
display(red_channel)
plt.imshow(red_channel)
```

```python
# For Blue Channel

blue_channel = image_array_colored[:, :, 2]
print(blue_channel)
display(blue_channel)
plt.imshow(blue_channel)
```

```python
# Extracting colored channels using pil

# Accessing individual channels
# Get the R, G, B, and A channels
r, g, b, a = image_colored.split()

display(r)
display(g)
display(b)
display(a)
```

```python
# Exttracting sub images

# Accessing a specific row - 90th row
row_90 = image_array_colored[90, :, :]
display(row_90)
plt.imshow(row_90)

```

```python
# accessing a specific column - 20th col
col_20 = image_array_colored[:, 20, :]
display(col_20)
plt.imshow(col_20)
```

```python
#Cropping image using PIL

# define the cropping box (left, upper, right, lower)
left = 100
upper = 50
right = 200
lower = 150

# Crop the image
cropped_image = image_colored.crop((left, upper, right, lower))

display(cropped_image)
```

```python
# Saving numpy array to image

# Create a numpy array
image_array_colored = np.zeros((100, 100, 3), dtype = np.uint8)

# Convert the numpy array back to a pillow image object
image_from_array = Image.fromarray(image_array_colored)

# display
image_from_array.show()

# Save
image_from_array.save("new_image.png")
```
