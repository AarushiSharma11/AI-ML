```python
def temperature_converter():
    """
    Converts temperature between Celsius and Fahrenheit.

    This function prompts the user to specify the conversion type
    (Celsius to Fahrenheit or Fahrenheit to Celsius) and then asks
    for the temperature value. It calculates and returns the
    converted temperature.

    Returns:
        float: The converted temperature value.
    """

    print("Choose conversion type:")
    print("1. Celsius to Fahrenheit")
    print("2. Fahrenheit to Celsius")

    # Get conversion choice from user
    choice = input("Enter 1 or 2: ")

    if choice == "1":
        # Celsius to Fahrenheit conversion
        celsius = float(input("Enter temperature in Celsius: "))
        fahrenheit = (celsius * 9/5) + 32
        print(f"{celsius}°C is equal to {fahrenheit}°F")
        return fahrenheit

    elif choice == "2":
        # Fahrenheit to Celsius conversion
        fahrenheit = float(input("Enter temperature in Fahrenheit: "))
        celsius = (fahrenheit - 32) * 5/9
        print(f"{fahrenheit}°F is equal to {celsius}°C")
        return celsius

    else:
        print("Invalid choice. Please enter 1 or 2.")
        return None


# Call the function
if __name__ == "__main__":
    temperature_converter()
```

```python
# Task 1

def unit_converter():
    """
    Converts units of measurement for length, weight, and volume.

    The function asks the user to select the type of conversion and
    then enter a value. It performs the selected conversion and
    displays the result.

    Returns:
        float: The converted value.
    """

    print("Select the type of conversion:")
    print("1. Length")
    print("2. Weight")
    print("3. Volume")

    try:
        choice = int(input("Enter your choice (1-3): "))

        # Ask for value to convert
        value = float(input("Enter the value to convert: "))

        if choice == 1:
            print("Length Conversion Options:")
            print("1. Meters to Feet")
            print("2. Feet to Meters")
            option = int(input("Enter your choice (1-2): "))

            if option == 1:
                result = value * 3.28084
                print(f"{value} meters = {result:.2f} feet")
                return result
            elif option == 2:
                result = value / 3.28084
                print(f"{value} feet = {result:.2f} meters")
                return result
            else:
                print("Invalid length conversion option.")

        elif choice == 2:
            print("Weight Conversion Options:")
            print("1. Kilograms to Pounds")
            print("2. Pounds to Kilograms")
            option = int(input("Enter your choice (1-2): "))

            if option == 1:
                result = value * 2.20462
                print(f"{value} kg = {result:.2f} lbs")
                return result
            elif option == 2:
                result = value / 2.20462
                print(f"{value} lbs = {result:.2f} kg")
                return result
            else:
                print("Invalid weight conversion option.")

        elif choice == 3:
            print("Volume Conversion Options:")
            print("1. Liters to Gallons")
            print("2. Gallons to Liters")
            option = int(input("Enter your choice (1-2): "))

            if option == 1:
                result = value * 0.264172
                print(f"{value} L = {result:.2f} gallons")
                return result
            elif option == 2:
                result = value / 0.264172
                print(f"{value} gallons = {result:.2f} L")
                return result
            else:
                print("Invalid volume conversion option.")

        else:
            print("Invalid conversion type.")

    except ValueError:
        print("Error: Please enter a valid numeric value.")

    return None


# Call the function
unit_converter()
```

```python
# Task 2

def calculate_sum(numbers):
    """
    Calculates the sum of a list of numbers.

    Parameters:
        numbers (list): A list of numeric values.

    Returns:
        float: The sum of the numbers in the list.
    """
    return sum(numbers)


def calculate_average(numbers):
    """
    Calculates the average of a list of numbers.

    Parameters:
        numbers (list): A list of numeric values.

    Returns:
        float: The average of the numbers.
    """
    return sum(numbers) / len(numbers)


def find_maximum(numbers):
    """
    Finds the maximum number in a list.

    Parameters:
        numbers (list): A list of numeric values.

    Returns:
        float: The largest number in the list.
    """
    return max(numbers)


def find_minimum(numbers):
    """
    Finds the minimum number in a list.

    Parameters:
        numbers (list): A list of numeric values.

    Returns:
        float: The smallest number in the list.
    """
    return min(numbers)


def main():
    try:
        print("Choose an operation:")
        print("1. Sum")
        print("2. Average")
        print("3. Maximum")
        print("4. Minimum")

        choice = int(input("Enter your choice (1-4): "))

        # User enters numbers separated by space
        user_input = input("Enter numbers separated by spaces: ")

        # Convert input into a list of floats
        numbers = [float(num) for num in user_input.split()]

        if len(numbers) == 0:
            print("Error: The list cannot be empty.")
            return

        if choice == 1:
            result = calculate_sum(numbers)
            print("Sum:", result)

        elif choice == 2:
            result = calculate_average(numbers)
            print("Average:", result)

        elif choice == 3:
            result = find_maximum(numbers)
            print("Maximum:", result)

        elif choice == 4:
            result = find_minimum(numbers)
            print("Minimum:", result)

        else:
            print("Invalid operation choice.")

    except ValueError:
        print("Error: Please enter valid numeric values.")


# Run the program
main()
```

```python
# Task 1 Flatten a Nested List:

def flatten(lst):
    """
    Flattens a nested list into a single list.

    Parameters:
        lst (list): A nested list.

    Returns:
        list: A flattened list.
    """
    result = []
    for item in lst:
        if isinstance(item, list):
            result.extend(flatten(item))
        else:
            result.append(item)
    return result


# Example
print(flatten([[1, 2], [3, 4], [5]]))
```

```python
# Task 2 Accessing Nested List Elements

def access_nested_element(lst, indices):
    """
    Accesses an element from a nested list using a list of indices.

    Parameters:
        lst (list): Nested list.
        indices (list): List of indices to access the element.

    Returns:
        element: The element at the specified position.
    """
    element = lst
    for i in indices:
        element = element[i]
    return element


# Example
lst = [[1,2,3],[4,5,6],[7,8,9]]
print(access_nested_element(lst, [1,2]))
```

```python
# Task 3 Sum of All Elements in a Nested List

def sum_nested(lst):
    """
    Calculates the sum of all numbers in a nested list.

    Parameters:
        lst (list): Nested list.

    Returns:
        int/float: Sum of all elements.
    """
    total = 0
    for item in lst:
        if isinstance(item, list):
            total += sum_nested(item)
        else:
            total += item
    return total


# Example
print(sum_nested([[1,2],[3,[4,5]],6]))
```

```python
# Task 4 Remove Specific Element from a Nested List

def remove_element(lst, elem):
    """
    Removes all occurrences of an element from a nested list.

    Parameters:
        lst (list): Nested list.
        elem: Element to remove.

    Returns:
        list: Modified list.
    """
    result = []
    for item in lst:
        if isinstance(item, list):
            result.append(remove_element(item, elem))
        elif item != elem:
            result.append(item)
    return result


# Example
print(remove_element([[1,2],[3,2],[4,5]],2))
```

```python
# Task 5 Find the Maximum Element in a Nested List

def find_max(lst):
    """
    Finds the maximum element in a nested list.

    Parameters:
        lst (list): Nested list.

    Returns:
        int/float: Maximum element.
    """
    max_value = float('-inf')

    for item in lst:
        if isinstance(item, list):
            max_value = max(max_value, find_max(item))
        else:
            max_value = max(max_value, item)

    return max_value


# Example
print(find_max([[1,2],[3,[4,5]],6]))
```

```python
def count_occurrences(lst, elem):
    """
    Counts how many times an element appears in a nested list.

    Parameters:
        lst (list): Nested list.
        elem: Element to count.

    Returns:
        int: Number of occurrences.
    """
    count = 0

    for item in lst:
        if isinstance(item, list):
            count += count_occurrences(item, elem)
        elif item == elem:
            count += 1

    return count


# Example
print(count_occurrences([[1,2],[2,3],[2,4]],2))
```

```python
# Task 7 Deep Flatten (Any Depth)

def deep_flatten(lst):
    """
    Flattens a deeply nested list into a single list.

    Parameters:
        lst (list): Deeply nested list.

    Returns:
        list: Flattened list.
    """
    result = []

    for item in lst:
        if isinstance(item, list):
            result.extend(deep_flatten(item))
        else:
            result.append(item)

    return result


# Example
print(deep_flatten([[[1,2],[3,4]],[[5,6],[7,8]]]))
```

```python
# Task 8 Nested List Average

def average_nested(lst):
    """
    Calculates the average of all elements in a nested list.

    Parameters:
        lst (list): Nested list.

    Returns:
        float: Average value.
    """
    flat = deep_flatten(lst)
    return sum(flat) / len(flat)


# Example
print(average_nested([[1,2],[3,4],[5,6]]))
```
